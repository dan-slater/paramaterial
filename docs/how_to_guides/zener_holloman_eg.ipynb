## Example of Zener-Holloman Analysis

```python
import paramaterial as pam
```

```python
ds = pam.DataSet('data/zh_eg_data', 'info/zh_eg_info.xlsx')
ds
```

```python
ds.info_table
```

```python
ds = ds.apply(pam.find_flow_stress, flow_strain=0.3)
df = ds.info_table
df['temperature_K'] = df['temperature'] + 273.15
df['rate_s-1'] = df['rate']
df['Q_activation'] = 156000
ds.info_table = df
ds.info_table
```

```python
ds = ds.apply(pam.calculate_ZH_parameter)
ds.info_table
```

```python
ds = pam.apply_ZH_regression(ds)
ds.info_table
```

```python
pam.plot_ZH_regression(ds)
```
